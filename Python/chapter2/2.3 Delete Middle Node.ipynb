# 2.3 Delete Middle Node

Implement an algorithm to delete a node in the middle (i.e any node but the first and last node, not necesarrily the exact middle of a singly linked list, given only access to that node)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
    def getData(self):
        return self.data
    
    def getNext(self):
        return self.next
        
    def setData(self,data):
        self.data = data
        
    def setNext(self, data):
        self.next = data
        
    def __str__(self):
        return str(self.data)

class LinkedList:
    def __init__(self, values=None):
        self.head = None
            
    def __iter__(self):
        current = self.head
        while current:
            yield current
            current = current.next
            
    def __str__(self):
        values = [str(x) for x in self]
        return ' -> '.join(values)
    
    def __len__(self):
        current = self.head
        count = 0
        while current.next != None:
            count += 1
            current = current.getNext()
        return count
    
    def add(self,item):
        temp = Node(item)
        temp.setNext(self.head)
        self.head = temp
        return temp
        
    def add_multiple(self,items):
        for item in items:
            self.add(item)

    def search(self, item):
        current = self.head
        while current != None:
            if current.getData() == item:
                return True
            current = current.getNext()
        return False
    
    def remove(self,item):
        current = self.head
        while current.getNext() != None:
            if current.getNext().getData() == item:
                current.setNext(current.getNext().getNext())
                break
            current = current.getNext()      
            
    def remove_duplicates(self):
        if self.head is None: return
        d = {}
        current = self.head
        
        while current.getNext() != None:
            value = current.getNext().getData()
            if value not in d:
                d[value] = True
                current = current.getNext()
            else:
                if current.getNext().getNext() is not None:
                    current.setNext(current.getNext().getNext())
                else:
                    current.setNext(None)
        return self
    
    def kth_to_last(self,k):
        runner = current = self.head
        
        for i in range(k):
            if runner is None:
                return None
            runner = runner.next
        
        while runner:
            current = current.next
            runner = runner.next
            
        return current
    
def delete_middle_node(node):
    """Problem being solved for exercise 2.3 """
    if node is None or node.next is None: return

    node.data = node.next.data
    node.next = node.next.next


                
    
```

```python
ll = LinkedList()
ll.add_multiple([1,2,3,4])
middle_node = ll.add(5)
ll.add_multiple([6,7,8])

print(ll)
delete_middle_node(middle_node)
print(ll)


```
